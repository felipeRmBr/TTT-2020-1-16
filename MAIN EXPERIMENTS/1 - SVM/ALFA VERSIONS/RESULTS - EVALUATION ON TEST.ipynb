```
import pickle
def loadResultsFiles(task):
  results_file = f'./Results/final/SVC-LINEAR_{task}_TEST.df'

  with open(results_file, 'rb') as file_handler:
    linear_results = pickle.load(file_handler)

  results_file = f'./Results/final/SVC-SIGMOID_{task}_TEST.df'

  with open(results_file, 'rb') as file_handler:
    sigmoid_results = pickle.load(file_handler)

  return linear_results, sigmoid_results
```
