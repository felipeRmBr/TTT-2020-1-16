## getBestModelResults()

```python
def getBestModelResults(task, architecture, type_list=['G']):
  if task=='HTA':
    sorting_metric='EMR'
  else:
    sorting_metric='f1-macro'

  results_list = [f'./Results/final/{architecture}-1_{task}_TEST.df',
                  f'./Results/final/{architecture}-2_{task}_TEST.df']

  results_df = results_utils.mergeDataFrames(results_list, '.delete_this.df') 

  type_mask = results_df.model_type.isin(type_list)
  best_model_info = results_df.loc[type_mask].sort_values(by=sorting_metric, ascending=False).iloc[0]
  best_model_id = best_model_info.conf_id
  print('TEST RESULTS')
  print(best_model_info)


  results_list = [f'./Results/final/{architecture}-1_{task}_VALIDATION.df',
                  f'./Results/final/{architecture}-2_{task}_VALIDATION.df'] 

  results_df = results_utils.mergeDataFrames(results_list, '.delete_this.df') 

  print('\nVALIDATION RESULTS')
  print(results_df.loc[results_df.conf_id==best_model_id].mean(axis=0))
  
```
