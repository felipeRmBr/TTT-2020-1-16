# Dependencies

```python
import json 
import numpy as np
import pandas as pd
import re
```

```python
# load dataset into memory
# import json
dataset = []
    
path_1 = 'train_es_A.json'
path_2 = 'dev_es_A.json'

for line in open(path_1, 'r'):
    dataset.append(json.loads(line))
    
for line in open(path_2, 'r'):
    dataset.append(json.loads(line))
```

```python
user_names = []
hashtags = []
urls = []
for entry in dataset:
    tweet = entry['text']
    user_names.extend(re.findall(r"@\S+", tweet))
    hashtags.extend(re.findall(r"#\S+", tweet))
    urls.extend(re.findall(r"https?://\S+", tweet))
```

### A first conclusion about user_names, hashtags and URLs

Given the findings presented above (i.e. that most of the usernames, hashtags and url's are indeed very unfrequent), it is fair to conclude that user_names, hashtags and url's are probably not that informative when diciding the label of a new tweet. Thus, it is reasonable to take one of the next two approaches when dealing with any of these elements:

    1) delete: replace every user_name, hashtag or url with an empty string ("")
    2) representative tagging: replace every user_name with the label _usr, every hashtag with the label _htg, and every url with the label _url
    
Either of this two approaches can be achieved using [regular expressions](https://docs.python.org/3.8/howto/regex.html) (see examples below)

## approach 1: deleting 

##### original tweet:

```python
t2
```

##### tweet with usr_names, hashtags and url's deleted:

```python
re.sub(r"@\S+|#\S+|https?://\S+", "", t2)
```

##### tweet with representative tagging

# Punctuation marks, numbers and other symbols (#, $, %, =, /, *, -, | ) 

Asssessing how helpfull are punctuation marks, numeric elements, and 'other symbols' is more tricky than it was with user_names, hashtags and url's. However, we'll define three approaches for dealing with these elements.

    1) keep it as it is (do nothing)
    2) delete all punctuation marks and numeric elements
    3) selective keeping

Approaches 2 and 3 are further explained in the following cells

## approach 2: delete all accurrences

Deleting all ocurrences of punctuations marks, numeric elements, and "other simbols" means that we only keep the [spanish] alphabetic caracters and spaces. This can be done using regular expressions.   

##### original tweet

```python
t3
```

##### puntuation marks, numeric elements, and other symbos deleted

```python
re.sub(r"[^A-za-z\sÁÉÍÓÚáéíóúÑñ]", '', t3)
```

## approach 3: selective keeping

For this approach we define a set of meaningful punctuation marks to be preserved along with the spaces and alphabetic characters.

    ['?', '!', '.', ',']
    
Once again, this can be done using regular expressions.

##### original tweet

```python
test_1 = "¿Es esto una #pregunta?, ¡¡Esto es una exclamación con varios símbolos adicionales ($ % &)!!, Y esta es una oración que termina con un punto."
test_1
```

##### preprocessed tweet with selective keeping

```python
re.sub(r"[^A-za-z\sÁÉÍÓÚáéíóúÑñ'?!\.,\d]", '', test_1)
```

# Complete preprocessing

The complete chain of preprocessing operations for a document would work as follows:

1. User-name, hashtag, and URLs handling.
2. Punctuation marks, numeric elements and other-symbols handling
3. Lowercasing
4. Tokenization

See the **example below**:

### original tweet

```python
t3
```

### preprocessing

```python
# represetative tagging for user_names, hashtags and URLs
t3_ = re.sub(r"@\S+", "_usr", t3)
t3_ = re.sub(r"#\S+", "_htg", t3_)
t3_ = re.sub(r"https?://\S+", "_url", t3_)
```

```python
# handling of punctuation marks and numerical elements
t3_ = re.sub(r"[^A-za-z\sÁÉÍÓÚáéíóúÑñ'?!\.,\d]", '', t3_)
```

```python
# lowercasing
t3_ = t3_.lower()
```

```python
# tokenization
t3_ = word_tokenize(t3_)
```

## Stemming

In order to normalize the documents even further, we can try running some [stemming](https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html) algorithm over the list of words that we get after tokenization. For this we can use the [Snowball](https://www.nltk.org/howto/stem.html) implementation provided with NLTK.

```python
from nltk.stem.snowball import SnowballStemmer
```

The SnowBall stemmer is avilable for several languages

```python
print(", ".join(SnowballStemmer.languages))
```

```python
stemmer = SnowballStemmer("spanish")
```

```python
testing_set = ['chiba', 'chica', 'chicago', 'chicana', 'chicano', 'chicas', 'chicharrones', 'chichimecas', 'chicles', 'chico']

for w in testing_set:
    print(stemmer.stem(w))
```
